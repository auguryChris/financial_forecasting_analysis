# Tuned XGBoost Regression Analysis

```python
import sys
sys.path.append('../../scripts')

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
from xgboost_for_tickers import TickerXGBRegressor
import pandas as pd
import numpy as np
```

```python
def preprocess(data, nonfeatures=['unnamed: 0', 'dimension', 'calendardate', 'datekey', 'lastupdated']):
    # keep feature data only
    data.rename({'reportperiod':'timestamp'}, inplace=True, axis=1)
    data.columns = [c.lower().strip() for c in data.columns]
    data.drop(nonfeatures, axis=1, inplace=True)
    return data
```

```python
ticker_model_dict = {'ticker_data':[], 'models':[]}
for data_name in ['MSFT', 'HD', 'UNH', 'XOM', 'ADSK', 'WAT']:
    ticker_model_dict['ticker_data'].append(preprocess(pd.read_csv(f'../../data/ticker_data/{data_name}_full_data.csv')))
```

```python
p1 = {'colsample_bytree': 0.75, 'gamma': 0, 'learning_rate': 0.01, 'max_depth': 4, 'n_estimators': 200, 'objective': 'reg:squarederror'}

p2 = {'colsample_bytree': 0.75, 'gamma': 0.1, 'learning_rate': 0.005, 'max_depth': 2, 'n_estimators': 750, 'objective': 'reg:squarederror'}

p3 = {'colsample_bytree': 0.5, 'gamma': 0, 'learning_rate': 0.01, 'max_depth': 2, 'n_estimators': 500, 'objective': 'reg:squarederror'}

p4 = {'colsample_bytree': 0.75, 'gamma': 0.1, 'learning_rate': 0.01, 'max_depth': 2, 'n_estimators': 750, 'objective': 'reg:squarederror'}

p5 = {'colsample_bytree': 0.5, 'gamma': 0.1, 'learning_rate': 0.01, 'max_depth': 4, 'n_estimators': 750, 'objective': 'reg:squarederror'}

p6 = {'colsample_bytree': 0.5, 'gamma': 0.1, 'learning_rate': 0.005, 'max_depth': 4, 'n_estimators': 750, 'objective': 'reg:squarederror'}
```

```python
param_list = [p1,p2,p3,p4,p5,p6]
for p in param_list:
    for key, val in p.items():
        p[key] = [val]
```

```python
i=0
for df in ticker_model_dict['ticker_data']:
    modler = TickerXGBRegressor(df.copy())
    modler.predict_with_xgb(target='high', gsearch_params=param_list[i],
    normilization_type='log_return', test_size=0.2, timeframe=-1, scale_type='minmax',  get_importances=True)
    
    ticker_model_dict['models'].append(modler)
    i+=1
```

```python
tickers = ['MSFT', 'HD', 'UNH', 'XOM', 'ADSK', 'WAT']
```

Note: There was a direct relationship between a greater holdout (test) r^2 normalized and being more profitable on the backtest

```python
for i in range(len(tickers)):
    print(tickers[i])
    print(ticker_model_dict['models'][i].holdout_metrics)
```

#### Feature importance averages across all tickers

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
avg_importances = {}
for i in range(len(tickers)):
    ticker_importance = ticker_model_dict['models'][i].feature_importance_dict
    for key, val in ticker_importance.items():
        # add key to dict if not added yet
        if key not in avg_importances.keys():
            avg_importances[key] = 0
        avg_importances[key] += ticker_importance[key]
```

```python
for key, val in avg_importances.copy().items():
    if avg_importances[key] <= 0:
        avg_importances.pop(key, None)
    else:
        avg_importances[key] = avg_importances[key] / len(tickers)
```

```python
avg_importances = dict(sorted(avg_importances.items(), key=lambda items: items[1], reverse=True))
```

```python
importance_df = pd.DataFrame(avg_importances, index=[0])
# For top 50 importances
importance_df = importance_df.iloc[:, 0:51]
```

#### Plotting Top 50 importances 

```python
# X axis labels and toggles to put into the bar chart code
Xlabels = list(importance_df.columns)
bar_width = 0.25
opacity = 0.8

# height of the bars for y-axis
bars1 = importance_df.values[0]

# set position of bar on x-axis
r1 = np.arange(len(bars1))

# create the plot and set the size
fig, ax = plt.subplots(figsize=(20,8))

rect1 = ax.bar(r1,bars1, alpha=opacity)

# x-axis - set tick labels and rotate the labels
plt.xticks(np.arange(0,len(Xlabels)), Xlabels)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontSize = 14)
ax.set_xlim(left = -0.25, right = len(Xlabels))

# y-axis - set format and add an axis label
#ax.set_ylim(bottom = 0, top = 4)
plt.ylabel("Absolute Tree Shap Value (Feature Importance Rating)", fontSize = 14)
# ax.get_yaxis().set_major_formatter(
#     ticker.FuncFormatter(lambda x, p: format(float(x), ',')))
ax.yaxis.grid(alpha = 0.5)

# add chart color scheme, title, and legend
plt.style.use('fivethirtyeight')  #('seaborn-colorblind')
plt.title("Average Features Importance Rating for XGBoosted Trees", fontSize = 20)  

#show chart
plt.show();
```
