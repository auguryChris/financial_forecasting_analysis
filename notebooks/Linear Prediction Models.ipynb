```python
import warnings
import pandas as pd
from sklearn.exceptions import ConvergenceWarning
from utils import DataSplit, Strategy
```

```python
ticker_names = ['MSFT', 'HD', 'UNH', 'XOM', 'ADSK', 'WAT']
```

```python
warnings.filterwarnings('ignore', category=pd.core.common.SettingWithCopyWarning)
warnings.filterwarnings('ignore', category=ConvergenceWarning)

# Assign each ticker to strategy object
msft = Strategy('MSFT', original_target_name='High',num_forecasts='all')
hd = Strategy('HD', original_target_name='High',num_forecasts='all')
unh = Strategy('UNH', original_target_name='High',num_forecasts='all')
xom = Strategy('XOM', original_target_name='High',num_forecasts='all')
adsk = Strategy('ADSK', original_target_name='High',num_forecasts='all')
wat = Strategy('WAT', original_target_name='High',num_forecasts='all')
```

```python
ticker_objects = [msft, hd, unh, xom, adsk, wat]
```

```python
hd.ticker
```

```python
# Results by model
ar_res = {'ticker' : [], 'mse' : [], 'mae' : [], 'mape' : [], 'rmse' : []}
arima_res = {'ticker' : [], 'mse' : [], 'mae' : [], 'mape' : [], 'rmse' : []}
linreg_res = {'ticker' : [], 'mse' : [], 'mae' : [], 'mape' : [], 'rmse' : []}
rf_res = {'ticker' : [], 'mse' : [], 'mae' : [], 'mape' : [], 'rmse' : []}

for t in ticker_objects:
    
    # Append ticker
    ticker = t.ticker
    ar_res['ticker'].append(ticker)
    arima_res['ticker'].append(ticker)
    linreg_res['ticker'].append(ticker)
    rf_res['ticker'].append(ticker)
    
    # Append MSE
    ar_res['mse'].append(t.ar_results['mse'])
    arima_res['mse'].append(t.arima_results['mse'])
    linreg_res['mse'].append(t.linreg_results['mse'])
    rf_res['mse'].append(t.rf_reg_results['mse'])
    
    # Append MAE
    ar_res['mae'].append(t.ar_results['mae'])
    arima_res['mae'].append(t.arima_results['mae'])
    linreg_res['mae'].append(t.linreg_results['mae'])
    rf_res['mae'].append(t.rf_reg_results['mae'])
    
    # Append MAPE
    ar_res['mape'].append(t.ar_results['mape'])
    arima_res['mape'].append(t.arima_results['mape'])
    linreg_res['mape'].append(t.linreg_results['mape'])
    rf_res['mape'].append(t.rf_reg_results['mape'])
    
    # Append RMSE
    ar_res['rmse'].append(t.ar_results['rmse'])
    arima_res['rmse'].append(t.arima_results['rmse'])
    linreg_res['rmse'].append(t.linreg_results['rmse'])
    rf_res['rmse'].append(t.rf_reg_results['rmse'])
    
    
ar_res = pd.DataFrame(ar_res).set_index('ticker')
arima_res = pd.DataFrame(arima_res).set_index('ticker')
linreg_res = pd.DataFrame(linreg_res).set_index('ticker')
rf_res = pd.DataFrame(rf_res).set_index('ticker')
```

```python
# Results by ticker
for t in ticker_objects:
    
    df = {'model' : [], 'mse' : [], 'mae' : [], 'mape' : [], 'rmse' : []}
    
    df['model'].extend(['AR', 'ARIMA', 'Linear Regression', 'Random Forest Regression'])
    df['mse'].extend([t.ar_results['mse'], t.arima_results['mse'], t.linreg_results['mse'], t.rf_reg_results['mse']])
    df['mae'].extend([t.ar_results['mae'], t.arima_results['mae'], t.linreg_results['mae'], t.rf_reg_results['mae']])
    df['mape'].extend([t.ar_results['mape'], t.arima_results['mape'], t.linreg_results['mape'], t.rf_reg_results['mape']])
    df['rmse'].extend([t.ar_results['rmse'], t.arima_results['rmse'], t.linreg_results['rmse'], t.rf_reg_results['rmse']])
    
    df = pd.DataFrame(df).set_index('model')
    
    print('\n' + t.ticker)
    print(df)
    
```
