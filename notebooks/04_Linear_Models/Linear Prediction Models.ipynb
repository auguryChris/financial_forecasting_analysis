```python
import warnings
import pandas as pd
from sklearn.exceptions import ConvergenceWarning
from utils import DataSplit, Strategy
```

```python
ticker_names = ['MSFT', 'HD', 'UNH', 'XOM', 'ADSK', 'WAT']
```

```python
# THIS CELL TAKES A WHILE TO RUN

warnings.filterwarnings('ignore', category=pd.core.common.SettingWithCopyWarning)
warnings.filterwarnings('ignore', category=ConvergenceWarning)

# Assign each ticker to strategy object
msft = Strategy('MSFT', original_target_name='High',num_forecasts='all')
hd = Strategy('HD', original_target_name='High',num_forecasts='all')
unh = Strategy('UNH', original_target_name='High',num_forecasts='all')
xom = Strategy('XOM', original_target_name='High',num_forecasts='all')
adsk = Strategy('ADSK', original_target_name='High',num_forecasts='all')
wat = Strategy('WAT', original_target_name='High',num_forecasts='all')
```

```python
ticker_objects = [msft, hd, unh, xom, adsk, wat]
```

```python
# Results by model
ar_res = {'ticker' : [], 'mse' : [], 'mae' : [], 'mape' : [], 'rmse' : []}
arima_res = {'ticker' : [], 'mse' : [], 'mae' : [], 'mape' : [], 'rmse' : []}
linreg_res = {'ticker' : [], 'mse' : [], 'mae' : [], 'mape' : [], 'rmse' : []}
rf_res = {'ticker' : [], 'mse' : [], 'mae' : [], 'mape' : [], 'rmse' : []}

for t in ticker_objects:
    
    # Append ticker
    ticker = t.ticker
    ar_res['ticker'].append(ticker)
    arima_res['ticker'].append(ticker)
    linreg_res['ticker'].append(ticker)
    rf_res['ticker'].append(ticker)
    
    # Append MSE
    ar_res['mse'].append(t.ar_test_results['mse'])
    arima_res['mse'].append(t.arima_test_results['mse'])
    linreg_res['mse'].append(t.linreg_test_results['mse'])
    rf_res['mse'].append(t.rfreg_test_results['mse'])
    
    # Append MAE
    ar_res['mae'].append(t.ar_test_results['mae'])
    arima_res['mae'].append(t.arima_test_results['mae'])
    linreg_res['mae'].append(t.linreg_test_results['mae'])
    rf_res['mae'].append(t.rfreg_test_results['mae'])
    
    # Append MAPE
    ar_res['mape'].append(t.ar_test_results['mape'])
    arima_res['mape'].append(t.arima_test_results['mape'])
    linreg_res['mape'].append(t.linreg_test_results['mape'])
    rf_res['mape'].append(t.rfreg_test_results['mape'])
    
    # Append RMSE
    ar_res['rmse'].append(t.ar_test_results['rmse'])
    arima_res['rmse'].append(t.arima_test_results['rmse'])
    linreg_res['rmse'].append(t.linreg_test_results['rmse'])
    rf_res['rmse'].append(t.rfreg_test_results['rmse'])
    
    
ar_res = pd.DataFrame(ar_res).set_index('ticker')
arima_res = pd.DataFrame(arima_res).set_index('ticker')
linreg_res = pd.DataFrame(linreg_res).set_index('ticker')
rf_res = pd.DataFrame(rf_res).set_index('ticker')
```

```python
# Results by ticker
for t in ticker_objects:
    
    df = {'model' : [], 'mse' : [], 'mae' : [], 'mape' : [], 'rmse' : []}
    
    df['model'].extend(['AR', 'ARIMA', 'Linear Regression', 'Random Forest Regression'])
    df['mse'].extend([t.ar_test_results['mse'], t.arima_test_results['mse'], t.linreg_test_results['mse'], t.rfreg_test_results['mse']])
    df['mae'].extend([t.ar_test_results['mae'], t.arima_test_results['mae'], t.linreg_test_results['mae'], t.rfreg_test_results['mae']])
    df['mape'].extend([t.ar_test_results['mape'], t.arima_test_results['mape'], t.linreg_test_results['mape'], t.rfreg_test_results['mape']])
    df['rmse'].extend([t.ar_test_results['rmse'], t.arima_test_results['rmse'], t.linreg_test_results['rmse'], t.rfreg_test_results['rmse']])
    
    df = pd.DataFrame(df).set_index('model')
    
    print('\n' + t.ticker)
    print(df)
    
```

## Prediction Data to CSV for Backtesting

Test Set Results to CSV (9/2019 - 11/2021)

```python
for model in ticker_objects:
    t = model.ticker
    print(f'Loading prediction data for {t}.')
    
    # Save AR predictions CSV
    ar_df = model.ar_test_forecasts
    ar_df = ar_df.to_frame()
    ar_df = ar_df.reset_index()
    ar_df.rename(columns={'index' : 'timestamp'}, inplace=True)
    ar_df['timestamp'] = pd.to_datetime(ar_df['timestamp'])
    ar_df.to_csv(f'../../data/ticker_predictions/ar_preds/{t}_ar_preds.csv')


    # Save ARIMA predictions CSV
    arima_df = model.arima_test_forecasts
    arima_df = arima_df.to_frame()
    arima_df = arima_df.reset_index()
    arima_df.rename(columns={'index' : 'timestamp'}, inplace=True)
#     arima_df['reportperiod'] = arima_df['reportperiod'].to_timestamp()
    arima_df.to_csv(f'../../data/ticker_predictions/arima_preds/{t}_arima_preds.csv')

    # Save Linear Regression predictions CSV
    linreg_df = model.linreg_test_forecasts
    linreg_df = linreg_df.to_frame()
    linreg_df = linreg_df.reset_index()
    linreg_df.rename(columns={'index' : 'timestamp'}, inplace=True)
    linreg_df['timestamp'] = pd.to_datetime(linreg_df['timestamp'])
    linreg_df.to_csv(f'../../data/ticker_predictions/linreg_preds/{t}_linreg_preds.csv')
    
    # Save Linear Regression predictions CSV
    rfreg_df = model.rfreg_test_forecasts
    rfreg_df = rfreg_df.to_frame()
    rfreg_df = rfreg_df.reset_index()
    rfreg_df.rename(columns={'index' : 'timestamp'}, inplace=True)
    rfreg_df['timestamp'] = pd.to_datetime(rfreg_df['timestamp'])
    rfreg_df.to_csv(f'../../data/ticker_predictions/rfreg_preds/{t}_rfreg_preds.csv')

    print(f'Prediction data for {t} saved.')
    
```
